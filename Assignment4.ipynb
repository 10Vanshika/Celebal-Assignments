 Load and Understand the Dataset


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")  # If using locally, use 'train.csv'
df.head()
```

Basic Info:


```python
df.info()
```

Summary Statistics & Distributions

```python
df.describe(include='all')
```

Numerical Feature Distributions

```python
num_cols = ['Age', 'Fare', 'SibSp', 'Parch']
df[num_cols].hist(bins=30, figsize=(10,6), layout=(2,2), color='skyblue')
plt.suptitle("Histograms of Numerical Features", fontsize=16)
plt.tight_layout()
```

Missing Value Analysis

```python
df.isnull().sum().sort_values(ascending=False)
```

Heatmap of Missing Data:

```python
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

Outlier Detection
 Boxplots for Outliers

```python
for col in ['Age', 'Fare']:
    sns.boxplot(data=df, x=col)
    plt.title(f'Boxplot of {col}')
    plt.show()
```

 Univariate Analysis
 Target Variable (Survived)

```python
sns.countplot(data=df, x='Survived')
plt.title("Survival Count")
```

Gender Distribution

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title("Survival by Gender")
```

Passenger Class

```python
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title("Survival by Passenger Class")
```

Bivariate & Multivariate Analysis
 Correlation Heatmap

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
```

 Age vs. Survival

```python
sns.violinplot(data=df, x='Survived', y='Age', inner='quartile')
plt.title("Age Distribution by Survival")
```

Fare vs. Survival

```python
sns.boxplot(data=df, x='Survived', y='Fare')
plt.title("Fare Distribution by Survival")
```

Categorical Variable Analysis
 SibSp and Parch

```python
sns.countplot(data=df, x='SibSp', hue='Survived')
plt.title("Survival by SibSp")

sns.countplot(data=df, x='Parch', hue='Survived')
plt.title("Survival by Parch")
```

Embarked Port

```python
sns.countplot(data=df, x='Embarked', hue='Survived')
plt.title("Survival by Embarked Port")
```
